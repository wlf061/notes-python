##合并数据集

###数据集的merge 和join

![avatar](file:///E:/study/mechain-learning/python-node/notes-python/12-pandas/5-10-5.png)

```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.DataFrame({'key':['b','b','c','c','a','d'],'data':[1,2,3,4,5,6]})
df1
```

```python
df2 = pd.DataFrame({'key':['b','c','a'],'data':[1,2,3]})
df2
```

```python
pd.merge(df1,df2, on='key')   ###merge默认为交集
```

```python
pd.merge(df1,df2, how='outer', on="key")    ###用how 指定连接方式
```

###索引上的合并

```python
left1 = pd.DataFrame({'key':['a','b','c','d'],'value':range(4)})
left1
```

```python
right1 = pd.DataFrame({'group_val':[3.5,7]},index=['a','b'])
right1
```

```python
pd.merge(left1,right1,left_on='key',right_index=True)
```

##轴向连接

![avatar](file:///E:/study/mechain-learning/python-node/notes-python/12-pandas/5-10-6.png)

```python
###concat 默认在axis =0 上进行连接，
s1=pd.Series([0,1], index=['a','b'])
s2=pd.Series([2,3,4], index=['c','d','e'])
s3=pd.Series([5,6],index=['f','g'])
pd.concat([s1,s2,s3])  
```

```python
pd.concat([s1,s2,s3],axis=1)  
```

```python
pd.concat([s1,s2,s3],axis=1, join='inner')  
```

```python
pd.concat([s1,s2,s3],keys=['one','two','three'] )   ###使用keys 创建层次化索引
```

```python
pd.concat([s1,s2,s3],axis=1,keys=['one','two','three'])   ###axis=1 时形成一个DataFrame
```
